```python
def build_date(date):
    output = str(date[0])

    if len(str(date[1])) == 1:
        output += str(0) 
    output += str(date[1])

    if len(str(date[2]))== 1:
        output += str(0)
    output += str(date[2])
    
    return int(output)
```

```python
date_list = []
ls = [d for d in df.date]

for d in ls:
    date = build_date(d)
    date_list.append(date)
```
